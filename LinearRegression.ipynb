```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
df.head()
```

```python
x = np.array(df.iloc[:,0], ndmin=2) # We denote X as Years of Experience
y = np.array(df.iloc[:,-1]) # We denote Y as Salary

x.shape, y.shape
```

```python
x = x.reshape(-1,1) # We reshape X to be a 2D array
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=1/3) # We split the data into training and testing sets

regressor = LinearRegression() # We create a Linear Regression object

regressor.fit(x_train,y_train) # We fit the training data to the regressor object
y_pred = regressor.predict(x_test)
```

```python
def mse(actual, predicted): # We define a function to calculate the mean squared error
    return np.mean(np.square(actual - predicted))
```

```python
y_test, y_pred
```

```python
mean_squared_error = mse(y_test, y_pred)
print("Mean Sqaured Error = ",(mean_squared_error))
print("Root Mean Sqaured Error(RMSE) =",(mean_squared_error**0.5))
```

```python
# Training Data plot

plt.scatter(x_train,y_train,color="red")
plt.plot(x_train,regressor.predict(x_train),color="blue")
plt.title('Salary vs Experience(Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Testing Data plot
plt.scatter(x_test,y_test,color='red')
plt.plot(x_test,y_pred,color='blue')
plt.title('Salary vs Experience(Testing set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
